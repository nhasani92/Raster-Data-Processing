```python
x = np.arange(0,10,0.2)
y_sin=np.sin(x)
y_cos=np.cos(x)
y_new=x*10
```

```python
plt.plot(x,y_sin,label='Sine', linewidth=4.0)
plt.plot(x,y_cos, label='Cos')
plt.plot(x,y_new, label='Cos')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
plt.close()
```
